Notebook to merge 2009 closure times and frequencies with 2019 income data.

```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
income2019blockgdf = gpd.read_file('incomeblocksbasemap.geojson')
income2019blockgdf.sample(5)
```

```python
income2019blockgdf['newgeoid'] = income2019blockgdf['geoid'].str[8:19]
income2019blockgdf.sample(5)
```

```python
CountClose2009gdf = gpd.read_file('2009_closetime_webmercator.geojson')
CountClose2009gdf.sample(5)
```

```python
CountClose2009gdf['newgeoid'] = CountClose2009gdf['geoid10'].str[1:12]
CountClose2009gdf.sample(5)
```

```python
check2gdf = CountClose2009gdf.newgeoid.value_counts()
check2gdf.sort_values()
```

```python
new2009df = CountClose2009gdf[['newgeoid', 'counts','counts_lag','counts_lag_diff','closure_median','closure_lag','closure_lag_diff']]
new2009df.sample(3)
```
