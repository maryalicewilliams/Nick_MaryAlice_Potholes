```python
import pandas as pd

import geopandas as gpd
 
import contextily as ctx

import matplotlib.pyplot as plt
```

```python
race2019blockgdf = gpd.read_file('race2019blocks.geojson')
race2019blockgdf.sample(3)
```

```python
race2019blockgdf['newgeoid'] = race2019blockgdf['geoid'].str[8:19]
race2019blockgdf.sample(3)
```

```python
checkgdf = race2019blockgdf.newgeoid.value_counts()
checkgdf.sort_values()
```

```python
CountClose2019gdf = gpd.read_file('CountClose2019.geojson')
CountClose2019gdf.sample(3)
```

```python
CountClose2019gdf['newgeoid'] = CountClose2019gdf['geoid10'].str[1:12]
CountClose2019gdf.sample(3)
```

```python
check2gdf = CountClose2019gdf.newgeoid.value_counts()
check2gdf.sort_values()
```

```python
new2019df = CountClose2019gdf[['newgeoid', 'counts','counts_lag','counts_lag_diff','closure_median','closure_lag','closure_lag_diff']]
new2019df.sample(3)
```

```python
Merged2019gdf = race2019blockgdf.merge(new2019df,on='newgeoid')
Merged2019gdf.sample(3)
```

```python
basic311gdf = gpd.read_file('Data/2019_OAK_311_NoBlank.csv')
basic311gdf.sample(3)
```

```python
basic311gdf['LAT'] = basic311gdf['LAT'].astype('float64')
```

```python
basic311gdf['LONG'] = basic311gdf['LONG'].astype('float64')
```

```python
race2019blockgdf = race2019blockgdf.to_crs(4326)
```

```python
joinedgdf = gpd.sjoin(basic311gdf, race2019blockgdf, how="left")
```

```python
pd.set_option('display.max_columns', None)
joinedgdf
```

```python
joinedgdf['newgeoid.value_counts()
```

```python
countdf = joinedgdf['newgeoid'].value_counts().rename_axis('newgeoid').reset_index(name='counts')
print (countdf)
```

```python
blockcountracegdf = race2019blockgdf.merge(countdf,on='newgeoid')
blockcountracegdf.sample(3)
```

```python
blockcountracegdf.drop(columns='geoid10')
```

```python
closeracegdf = joinedgdf[joinedgdf['Time to Close']!='Open']
closeracegdf.sample(3)
```

```python
closeracegdf['Time to Close'] = closeracegdf['Time to Close'].astype('int64')
```

```python
closeddf_grouped = closeracegdf.groupby(['newgeoid']).median()[['Time to Close']]
closeddf_grouped.sample(3)
```

```python
closetimedf = closedgdf_grouped.rename_axis('newgeoid').reset_index('newgeoid')
print (closetimedf)
```

```python
closetimedf.columns = ['newgeoid', 'closure_median']
closetimedf
```

```python
blockcloseracegdf = race2019blockgdf.merge(closetimedf,on='newgeoid')
blockcloseracegdf
```

```python
blockcloseracegdf.drop(columns='geoid10')
```

```python
blockcloseracegdf.to_file("blockcloserace2019gdf.geojson", driver='GeoJSON')
```

```python
blockcountracegdf.to_file("blockcountrace2019gdf.geojson", driver='GeoJSON')
```
